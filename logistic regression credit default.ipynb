```python
# standard modules we will always use
import pandas as pd
import numpy as np
import scipy as sp
import mglearn
from IPython.display import display
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("default.csv")
display(data)
```

```python
X = data[['LIMIT_BAL','SEX','EDUCATION','MARRIAGE','AGE','PAY_0','PAY_2','PAY_3','PAY_4','PAY_5','PAY_6',
 'BILL_AMT1','BILL_AMT2','BILL_AMT3','BILL_AMT4','BILL_AMT5','BILL_AMT6','PAY_AMT1','PAY_AMT2','PAY_AMT3',
 'PAY_AMT4','PAY_AMT5','PAY_AMT6']]
Y = data['default_next_mo']
```

```python
res = pd.scatter_matrix(X, c=Y, figsize = (20,20), marker ='v',
                       hist_kwds={'bins': 20}, s = 60, alpha = .8, cmap=mglearn.cm3)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y, random_state = 42) #random state is to validate the same sequence everytime you run
```

```python
logreg = logit().fit(X_train, y_train)
print("training set score: {:.3f}".format(logreg.score(X_train, y_train)))
print('Test set score: {:.3f}'.format(logreg.score(X_test,y_test)))
```

```python
logreg100 = logit(C = 100).fit(X_train, y_train)
print("training set score: {:.3f}".format(logreg100.score(X_train, y_train)))
print('Test set score: {:.3f}'.format(logreg100.score(X_test,y_test)))
print(logreg.coef_)
```

```python
# all of the logit models are the same despite the regularizer
logreg001 = logit(C = .001).fit(X_train, y_train)
print("training set score: {:.3f}".format(logreg001.score(X_train, y_train)))
print('Test set score: {:.3f}'.format(logreg001.score(X_test,y_test)))
```

```python
logreg10000 = logit(C = 10000).fit(X_train, y_train)
print("training set score: {:.3f}".format(logreg10000.score(X_train, y_train)))
print('Test set score: {:.3f}'.format(logreg10000.score(X_test,y_test)))
```

```python
plt.plot(logreg.coef_.T, 'o', label = "C = 1")
plt.xticks(range(X.shape[1]), col_names, rotation = 90 )
plt.hlines(0,0, X.shape[1])
plt.ylim(-1, 1)
plt.xlabel("Coefficient Index")
plt.ylabel('Coefficient Magnitude')
plt.legend()
```

```python
# if we want to use L1 regularizer
for C, marker in zip([.001, 100, 10000], ['o','^','v']):
    lr_l1 = logit(C=C, penalty = 'l1').fit(X_train, y_train)
    print("training accuracy for l1 logreg with C={:.3f}: {:.2f}".format(
        C, lr_l1.score(X_train, y_train)))
    print("test accuracy for l1 logreg with C={:.3f}: {:.2f}".format(
        C, lr_l1.score(X_test, y_test)))
    plt.plot(lr_l1.coef_.T, marker, label ="C ={:.3f}".format(C))
    
plt.xticks(range(X.shape[1]), col_names, rotation = 90 )
plt.hlines(0,0, X.shape[1])
plt.ylim(-1, 1)
plt.xlabel("Coefficient Index")
plt.ylabel('Coefficient Magnitude')
plt.legend()   
```
